# cars.csv와 brand.csv의 데이터 합치기

```python
merged_df = pd.merge(car_df, brand_df, left_on='brand', right_on='title', how='left')
```

```python
merged_df.drop(columns=['title_y'], inplace=True)
```

```python
merged_df.head()
```

```python
merged_df.rename(columns={'title_x': 'title'}, inplace=True)
```

# 결측치 처리하기

```python
merged_df[merged_df.duplicated()]
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
merged_df
```

```python
merged_df['Service history'] = merged_df['Service history'].fillna('Unknown')
```

```python
merged_df.isna().sum()
```

```python
merged_df['Doors'].fillna(merged_df['Doors'].median(), inplace = True)
```

```python
merged_df['Seats'].fillna(merged_df['Seats'].median(), inplace = True)
```

```python
merged_df['Engine'].fillna(merged_df['Engine'].median(), inplace = True)
```

```python
merged_df['Emission Class'].fillna(merged_df['Emission Class'].median(), inplace = True)
```

```python
merged_df.dropna(subset = ['Previous Owners'], inplace = True)
```

```python
merged_df.isna().sum()
```

### 결측치 결과
- 데이터의 양이 그리 크지 않기때문에 최대한 데이터를 가지고 가는 쪽으로 해서 진행하려고 했음
- 하지만 일단 중복값들은 제거해야 수월하게 진행할 수 있을 것 같아 중복값 제거
- 그리고 제공된 힌트를 가지고 먼저 해결,  Service history의 NaN값을 Unknown으로 채워주기(fillna() 사용)
- 나머지 NaN값이 존재하는 Engine, Doors, Seats, Emission Class, Previous Owners 결측치 해결
- Engine, Doors, Seats, Emission Class는 fillna()로 median으로 중앙값으로 변경
- Previous Owners는 고민을 많이 했다. 이전 소유자가 NaN이면 뭘까 생각하다가...그냥 없애버렸다,, 얘도 중앙값으로 하려다가 고민의 시간이 너무 오래되어 일단 제거하였음(나머지처럼 fillna()로 중앙값으로 변경하는 것은 나중에 해보기로 )
- 어찌됐든 결측치는 모두 제거를 완료하였음.

# Scaler 활용하기, PCA 주성분 분석

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
```

### StandardScaler

```python
ss = StandardScaler()
```

```python
ss_cols = ['Price', 'Mileage(miles)', 'Registration_Year', 'Previous Owners', 'Engine', 'Doors', 'Seats', 'Emission Class']
```

```python
ss_df = ss.fit_transform(merged_df[ss_cols])
```

```python
pd.DataFrame(ss_df, columns = ss_cols)
```

#### 기존 데이터의 정보 70% 이상을 가지는 수준에서 최소한의 주성분 추출 (PCA)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.7)
```

```python
pca.fit(ss_df)
```

```python
pd.DataFrame(pca.fit_transform(ss_df))
```

```python
(pca.explained_variance_ratio_).sum() 
```

###  RobustScaler

```python
rs = RobustScaler()
```

```python
rs_df = rs.fit_transform(merged_df[ss_cols])
```

```python
pd.DataFrame(rs_df, columns = ss_cols)
```

```python
pca.fit(rs_df)
```

```python
pd.DataFrame(pca.fit_transform(rs_df))
```

```python
(pca.explained_variance_ratio_).sum() 
```

### MinMaxScaler

```python
mm = MinMaxScaler()
```

```python
mm_df = mm.fit_transform(merged_df[ss_cols])
```

```python
pd.DataFrame(mm_df, columns = ss_cols)
```

```python
pca.fit(mm_df)
```

```python
pd.DataFrame(pca.fit_transform(mm_df))
```

```python
(pca.explained_variance_ratio_).sum() 
```
